``Importing the sklearn library``

```python
from sklearn.model_selection import train_test_split

x=df[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
y=df['Yearly Amount Spent']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
len(x_train),len(y_train) # Both the variable have same no. of data
```

## ``Training the Model``

```python
from sklearn.linear_model import LinearRegression

lm=LinearRegression()
lm.fit(x_train,y_train)
```

```python
lm.coef_
```

```python
cdf=pd.DataFrame(lm.coef_,x.columns,columns=['Coef'])
print(cdf)
```

## ``Predictions of the Model``

```python
predictions=lm.predict(x_test)
predictions
```

```python
sns.scatterplot(x=predictions,y=y_test)
plt.ylabel("Truth")
plt.xlabel("Predictions")
plt.title('Evaluation of LM Model')
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
import math
```

```python
print("Mean Absolute Error of :" , mean_absolute_error(y_test,predictions))
print("Mean Squared Error of :" , mean_squared_error(y_test,predictions))
print("RMSE :" ,math.sqrt( mean_squared_error(y_test,predictions)))
```

### ``Residuals``

```python
residuals = y_test-predictions
residuals
```

```python
sns.displot(residuals,bins=20,kde=True)
plt.show()
```

## ``Visualization of the residuals``

```python
import pylab
import scipy.stats as stats

stats.probplot(residuals,dist='norm',plot=pylab)
pylab.show()
```
